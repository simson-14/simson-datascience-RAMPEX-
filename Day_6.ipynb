# Inter Quartile Range (IQR):

IQR = Q3-Q1
Q1 = 25%
Q3 = 75%

```python
import numpy as np
data = [5, 10, 15, 20, 25, 30, 35, 40]
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
```

```python
import numpy as np
data = [10, 12, 14, 15, 16, 18, 100]
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print(lower)
print(upper)
outliers = [x for x in data if x < lower or x > upper]
print("Outliers:", outliers)
```
